```python
df = df.dropna()
```

```python
df = df.drop(columns = ['id'])
```

```python
num_features = df.drop(columns = ['Class'])
```

```python
#Detecting the outlier by IQR method
Q1 = num_features.quantile(0.25)
Q2 = num_features.quantile(0.75)

IQR = Q2 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q2 - 1.5*IQR

outliers = ((num_features < lower_bound) | (num_features > upper_bound)).sum()

print(outliers.sort_values(ascending=False))
```

```python
#Box plot
import matplotlib.pyplot  as plt
import seaborn as sns

num_features.plot(kind = 'box' , subplots=True ,layout = (5,6), figsize = (15,10), patch_artist = True)

plt.tight_layout()
plt.show()
```

```python
#Correlation matrix

corr_matrix = df.corr()

plt.figure(figsize=(18,12))

sns.heatmap(corr_matrix , annot = True , cmap = 'coolwarm', linewidth = 0.5)

plt.title('Heatmap of feature Correlation Matrix')

plt.show()
```

```python
# Splitting Data
X = num_features
y = df['Class']

print(f'Features shape : {X.shape}')
print(f'Target shape : {y.shape}')
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
```

```python
print(df['Class'].value_counts(normalize=True)*100)
```

```python
#Visulizing how many fraudlent cases
sns.countplot(x=df['Class'])
plt.title("Fraud vs. Non-Fraud Transactions")
plt.xlabel("Class (0 = Non-Fraud, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
#Train test split
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test  = train_test_split(X_scaled, y ,test_size = 0.2 , random_state = 42, stratify=y)
```

```python
print(f"Training set shape: {X_train.shape}, Test set shape: {X_test.shape}")
print(f"Fraud cases in training: {y_train.value_counts()}")
print(f"Fraud cases in test: {y_test.value_counts()}")
```

```python
#Balancing the data
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_balance , y_train_balance = smote.fit_resample(X_train, y_train)

print(y_train_balance.value_counts())
```

```python
#Training the logistic model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report , confusion_matrix

logreg = LogisticRegression(class_weight='balanced', random_state=42)

logreg.fit(X_train_balance, y_train_balance)
```

```python
#checking the accuracy
y_pred = logreg.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Fraud', 'Fraud'], yticklabels=['No Fraud', 'Fraud'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#Gradient boost tree
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

gb = GradientBoostingClassifier(random_state = 42)

param_grid = {
    'n_estimators' : [50, 100 , 200],
    'learning_rate' : [0.01 ,0.1 ,  0.2],
    'max_depth' : [3, 4, 5],
}

```

```python
#grid_search_cv

grid_search = GridSearchCV(estimator=gb , param_grid=param_grid , cv=3 , n_jobs=-1,verbose=2)

grid_search.fit(X_train_balance,y_train_balance)

best_params = grid_search.best_params_
print("best hyperparameters : ", best_params)

best_gb = grid_search.best_estimator_

y_pred_gb = best_gb.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print("Classification Report:\n", classification_report(y_test, y_pred_gb))
```

```python
cm_gb = confusion_matrix(y_test, y_pred_gb)
sns.heatmap(cm_gb, annot=True, fmt='d', cmap='Blues', xticklabels=['No Fraud', 'Fraud'], yticklabels=['No Fraud', 'Fraud'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Gradient Boosting Confusion Matrix")
plt.show()
```
